```python
import pandas as pd
import numpy as np

#EXTRACT_PROCESS
net_df=pd.read_csv("netflix_titles.csv")


#TRANSFORM_PROCESS
#Checking duplicates and missing values
net_df.duplicated().sum()
net_df.isnull().sum()

#Delete several column with tiny missing values
net_df.dropna(
     subset=['duration', 'rating', 'date_added'],
     inplace=True)

#Handling the missing value
net_df['director'].fillna('Unknown',inplace=True)
net_df['cast'].fillna('Unknown',inplace=True)
net_df['country'].fillna('Unknown',inplace=True)
net_df.isnull().sum()


#Change the data type
net_df['date_added']=pd.to_datetime(net_df['date_added'],errors='coerce')


#create new columns: month_added, year_added
net_df['month_added']=net_df['date_added'].dt.month.astype('Int64')

net_df['year_added']=net_df['date_added'].dt.year.astype('Int64')
net_df=net_df.drop(columns=['cast','description'], errors='ignore')
net_df=net_df.drop(columns=['listed_in'], errors='ignore')


#separate the 'duration'
net_df['duration_minute']=np.where(net_df['type']=='Movie',net_df['duration'].str.replace('min', '', regex= False), np.nan).astype(float)

net_df['duration_season']=np.where(net_df['type']=='TV Show', net_df['duration'].str.replace('Seasons', '', regex=False).str.replace('Season','', regex=False),np.nan).astype(float)



#Delete date_added and duration column
net_df=net_df.drop(columns=['date_added','duration'], errors='ignore')

net_df.head()

#LOAD_PROCESS
net_df.to_csv("netflix_cleaned_version.csv", index=False)
```
